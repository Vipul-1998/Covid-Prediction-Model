```python
#Function for plotting bar chart

def bar_chart(feature):
    discharge = train[train['Present Status']=='DISCHARGE'][feature].value_counts()
    UT = train[train['Present Status']=='UNDER TREATMENT'][feature].value_counts()
    death = train[train['Present Status']=='DEATH'][feature].value_counts()
    df = pd.DataFrame([discharge,UT,death])
    df.index = ['DISCHARGE','UNDER TREATMENT','DEATH']
    df.plot(kind='barh', stacked=True, figsize=(40,20))
```

```python
#Plotting bar chart

bar_chart('Gender')
plt.title("Present Status VS Gender")
plt.ylabel("Present Status")
plt.xlabel("Gender")
```

```python
#Importing Libraries

from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
#Importing Functions

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits = 10, shuffle = True, random_state=0)
```

```python
#Executing Random Forest Algorithm

clf = RandomForestClassifier(n_estimators=6)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
#Mean of all accuracy values

round(np.mean(score)*100, 2)
```
